```python
#제목 부여하기
plt.plot([1,4,7,9,12,15]) # 도면에 추가
plt.title ("plt.plot([1,4,7,9,12,15])") #제목달기
plt.show() # 화면에 시각화
```

```python
#x,y 전달('.')
plt.plot([1,2,3,4,5,7],[1,2,9,7,12,15]) # 도면에 추가(x)
plt.title ("plt.plot([1,2,3,4,5,7],[1,2,9,7,12,15])") #제목달기(y)
plt.show() # 화면에 시각화
```

#fmt

'[maker][line][color]'

```python
#마커 사용
plt.plot([1,2,3,4,5,7],[1,2,9,7,12,15],'.') # 도면에 추가(x)
plt.title ("plt.plot([1,2,3,4,5,7],[1,2,9,7,12,15]),'.'") #제목달기(y)
plt.show() # 화면에 시각화
```

```python
#마커 사용('o')
plt.plot([1,2,3,4,5,7],[1,2,9,7,12,15],'o') # 도면에 추가(x)
plt.title ("plt.plot([1,2,3,4,5,7],[1,2,9,7,12,15]),'o'") #제목달기(y)
plt.show() # 화면에 시각화
```

```python
#마커 사용('*')
plt.plot([1,2,3,4,5,7],[1,2,9,7,12,15],'*') # 도면에 추가(x)
plt.title ("plt.plot([1,2,3,4,5,7],[1,2,9,7,12,15]),'*'") #제목달기(y)
plt.show() # 화면에 시각화
```

```python
#마커 + 라인 사용(- ,  o-)
plt.plot([1,2,3,4,5,7],[1,2,9,7,12,15],'o-') # 도면에 추가(x)
plt.title ("plt.plot([1,2,3,4,5,7],[1,2,9,7,12,15]),'o-'") #제목달기(y)
plt.show() # 화면에 시각화
```

```python
#마커 + 라인 사용( '-',  'o:' )  
plt.plot([1,2,3,4,5,7],[1,2,9,7,12,15],'o:') # 도면에 추가(x)
plt.title ("plt.plot([1,2,3,4,5,7],[1,2,9,7,12,15]),'o:'") #제목달기(y)
plt.show() # 화면에 시각화
```

```python
#마커 + 라인 + 색상 사용( '-',  'o:r' )  
plt.plot([1,2,3,4,5,7],[1,2,9,7,12,15],'o:r') # 도면에 추가(x)
plt.title ("plt.plot([1,2,3,4,5,7],[1,2,9,7,12,15]),'o:r'") #제목달기(y)
plt.show() # 화면에 시각화
```

```python
#kwargs의 color
plt.plot([1,2,3,4,5,7],[1,2,9,7,12,15],'o:',color = 'skyblue') # 도면에 추가(x)
plt.title ("plt.plot([1,2,3,4,5,7],[1,2,9,7,12,15]),'o:r',color = 'skyblue'") #제목달기(y)
plt.show() # 화면에 시각화
```

```python
#kwargs의 color
plt.plot([1,2,3,4,5,7],[1,2,9,7,12,15],'o:',color = '#cabdf9') # 도면에 추가(x)
plt.title ("plt.plot([1,2,3,4,5,7],[1,2,9,7,12,15]),'o:r',color = '#cabdf9'") #제목달기(y)
plt.show() # 화면에 시각화
```

```python
#kwargs의 linewidth 라인두께 
plt.plot([1,2,3,4,5,7],[1,2,9,7,12,15],'o:',color = '#cabdf9', linewidth = 5) # 도면에 추가(x)
plt.title ("plt.plot([1,2,3,4,5,7],[1,2,9,7,12,15]),'o:r',color = '#cabdf9', linewidth = 5") #제목달기(y)
plt.show() # 화면에 시각화
```

```python
#데이터를 두개 이상 전달하기
plt.plot([1,2,3,4,5,7],[1,2,9,7,12,15],'bo', [1,2,3,4,5,7],[8,9,10,15,10,15],'r^') # 도면에 추가(x)
plt.title ("plt.plot([1,2,3,4,5,7],[1,2,9,7,12,15]),'bo',[1,2,3,4,5,7],[8,9,10,15,10,15],'r^'") #제목달기(y)
plt.show() # 화면에 시각화
```

```python
#데이터를 두개 이상 전달하기
plt.plot([1,2,3,4,5],[90,80,85,77,97],'bo')
plt.plot([1,2,3,4,5],[95,89,87,72,90],'r^')
plt.axis([0,6,0,100]) # axis (축)
plt.title ("scores") #제목달기(y)
plt.show() # 화면에 시각화
```

```python
#축 이름 지정하기
plt.plot([1,2,3,4,5],[90,80,85,77,97],'bo')
plt.plot([1,2,3,4,5],[95,89,87,72,90],'r.')
plt.axis([0,6,0,100]) # axis (축)
plt.xlabel ('number') 
plt.ylabel ('score')
plt.title ("scores") #제목달기(y)
plt.show() # 화면에 시각화
```

```python
#축 이름 지정하기
plt.plot([1,2,3,4,5],[90,80,85,77,97],'bo', label = 'class1')
plt.plot([1,2,3,4,5],[95,89,87,72,90],'r.', label = 'class2')
plt.axis([0,6,0,100]) # axis (축)
plt.xlabel ('number') 
plt.ylabel ('score')
plt.title ("scores") #제목달기(y)
plt.legend()  #label를 표시하세요
plt.show() # 화면에 시각화
```

```python
#y 축에 표현할 값을 지정 + 디폴트 색상 확인
ys_str = []
for i in range(1,31): # 총 30개의 수평선 그리기 
  plt.plot([i,i])
  if i%3 == 0:
    ys_str.append(str(i))
  else:
    ys_str.append(" ")
plt.title("default color, set yticks")
plt.yticks(range(1,31), ys_str) # 색상을 지정해 주지 않아도 알아서 지정
plt.show()
```

#파이그래프

```python
locales = ['서울','강원','경북','충남','기타']
incomes = [2300,1200,1800,800,400]
plt.pie(incomes,labels = locales)
plt.title("지역별 수입")
plt.show()
```

```python
#돌출 시키기
locales = ['서울','강원','경북','충남','기타']
incomes = [2300,1200,1800,800,400]
exs = [0,0.1,0,0,0]
plt.pie(incomes,labels = locales, explode = exs)
plt.title("지역별 수입 -돌출 시키기")
plt.show()
```

```python
# 퍼센트 표시하기
locales = ['서울','강원','경북','충남','기타']
incomes = [2300,1200,1800,800,400]
exs = [0,0.1,0,0,0]
plt.pie(incomes,labels = locales, explode = exs, autopct = "%.2f")
plt.title("지역별 수입 - 퍼센트 시키기")
plt.show()
```

```python
# 퍼센트 표시할 위치 지정하기
locales = ['서울','강원','경북','충남','기타']
incomes = [2300,1200,1800,800,400]
exs = [0,0.1,0,0,0]
plt.pie(incomes,labels = locales, explode = exs, autopct = "%.2f",pctdistance=0.8) 
# pctdistance: 퍼센트를 겉에 가까이
plt.title("지역별 수입 - 퍼센트 표시할 위치 지정하기")
plt.show()
```

```python
#lable 위치 지정
locales = ['서울','강원','경북','충남','기타']
incomes = [2300,1200,1800,800,400]
exs = [0,0.1,0,0,0]
plt.pie(incomes,labels = locales, explode = exs, autopct = "%.2f",pctdistance=0.8, labeldistance=1.15)
plt.title("지역별 수입 - label 표시할 위치 지정하기")
plt.show()
```

```python
#시계 방향으로 각도 계산하기
locales = ['서울','강원','경북','충남','기타']
incomes = [2300,1200,1800,800,400]
exs = [0,0.1,0,0,0]
plt.pie(incomes,labels = locales, explode = exs, autopct = "%.2f",startangle=90, counterclock=False)
plt.title("지역별 수입 - 시계 방향으로 각도 계산하기")
plt.show()
```
