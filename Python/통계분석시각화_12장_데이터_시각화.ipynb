```python
x = np.arange(-4.5,5,0.5)
y1 = 2*x**2
y2 = 5*x + 30
y3 = 4*x**2+10
plt.plot(x,y1,'#abe7ea',label='y = 2*x^2')
plt.plot(x,y2,'r-',label = 'y = 5x+30')
plt.plot(x,y3,'g:', label = '4x^2+10')
plt.legend()
plt.show()
```

```python
plt.plot(x,y1,'bo',x,y2,'r-',x,y3,'g:')
plt.show()
```

```python
#figure를 설정하지 않을 때 디폴트는 figure(1)에 그림
plt.plot(x,y1)

plt.figure(2) #figure: 그래프들를 독립적으로 출력하려고 할 때 사용
plt.plot(x,y2)

plt.figure(3)
plt.plot(x,y3)
plt.show()
```

```python
x = np.arange(-4.5,5,0.5)
y = 2*x**3
plt.plot(x,y,label = 'data1')
plt.xlabel('x')
plt.ylabel('y')
plt.title("title, xlabel, ylabel")
plt.grid(True)
plt.legend()
plt.show()
```

```python
x = np.arange(-4.5,5,0.5)
y1 = 2*x**3
y2 = 3*x**2
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel('x')
plt.ylabel('y')
plt.title("title, xlabel, ylabel")
plt.grid(True)
plt.legend(['data1','data2'], loc = 'lower right')
plt.show()
```

#산점도

```python
height = [165,177,160,190,185,155,166]
weight = [62,67,55,87,66,77,66]

plt.plot(height,weight,'ro')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.grid(True)
plt.title("Height-Weight")
plt.show()
```

```python
plt.scatter(height,weight,s =200)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.grid(True)
plt.title("Height-Weight")
plt.show()
```

```python
size = 100 * np.arange(1,8)
colors = ['#b5e7f4','#abdce8','#8e8dd6','#4e88e0','#abe7ea','#0784f9','#121aed']
plt.scatter(height,weight,s = size, c=colors) # scatter: 산점도
plt.show()
```

```python
import numpy as np
city = ['seoul','incheon','daejeon','daegu','wolsan','busan','gangju']

lat = [37.56,37.45,36.35,35.87, 35.53,35.18,35.16]
lon = [126.97,126.70,127.38,128.60,129.31,129.07,126.85]
pop_den = [16154,2751,2839,2790,1099,4454,2995]
size = np.array(pop_den)*0.2
colors = ['#b5e7f4','#abdce8','#8e8dd6','#4e88e0','#abe7ea','#0784f9','#121aed']
plt.figure(figsize =(8,8))
plt.scatter(lon,lat,s=size, c =colors, alpha=0.5)
plt.xlabel("longfitude")
plt.ylabel("latitude")
plt.title("population density by region(2017)")
for x,y,name in zip(lon,lat,city):
  plt.annotate(name,(x,y)) #plt.text(x,y,name)
plt.show()
```

#막대 그래프

```python
ids =['m01','m02','m03','m04']
be_ex = [27,35,40,33]
af_ex = [30,38,42,37]
n = len(ids)
index = np.arange(n) #np.arra, [0,1,2,3]
plt.bar(ids, be_ex)
plt.grid(True)
plt.show()
```

```python
plt.bar(index,be_ex,tick_label = ids)
plt.show()
```

```python
colors = ['#b5e7f4','#abdce8','#8e8dd6','#4e88e0']
plt.bar(index,be_ex,color = colors, width = 0.6)
plt.show()
```

```python
plt.barh(index,be_ex,color = colors, height= 0.6)
plt.show()
```

```python
bw = 0.4
plt.bar(index,be_ex,color = '#abdce8', align = 'edge',width= bw,label='before')
plt.bar(index+bw,af_ex,color = '#8e8dd6', align = 'edge',width= bw,label='alter')
plt.xticks(index+bw,ids)
plt.title(" before/ after counting sit_up work out")
plt.legend()
plt.show()
```
